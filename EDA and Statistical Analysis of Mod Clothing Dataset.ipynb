```python
# Test Statistic(z_statistic)

def TwoSampZTest(samp_mean_1, samp_mean_2, samp_std_1, samp_std_2, value, n1, n2):
    denominator = np.sqrt((samp_std_1**2 / n1) + (samp_std_2**2 / n2))
    zscore = ((samp_mean_1 - samp_mean_2) - (value)) / denominator
    return zscore

# Given values :
nc_bar = 36           # samp_mean_1
nc_sigma = 7.6        # samp_std_1
sm_bar = 34           # samp_mean_2
sm_sigma = 5.8        # samp_std_2
sample_size = 100     # n1 & n2
value = 0             # value(Hypothesized Value)

z_statistic = round(TwoSampZTest(nc_bar,sm_bar,nc_sigma,sm_sigma,value,sample_size,sample_size),3)

p_val = round(stats.norm.sf(z_statistic),3)

result = pd.DataFrame({'Z_Critical':z_critical,'Z_Statistic':z_statistic,'Alpha':alpha,'P-value':p_val}, index=['Values'])
result
```

```python
table = pd.crosstab(df_new['category'], df_new['review_summary'])
observed_value = table.values
observed_value.shape
```

```python
### Step 1 : State the Hypotheses
# Ho : Observed frequency and Expected frequency have same distribution
# Ha : Observed frequency and Expected frequency have different distribution

### Step 2 : Decision Criteria
# Right Tail Test 
alpha = 0.05
r=7
c=26864
chisq_alpha = stats.chi2.isf(alpha,df=((r-1)*(c-1)))
print('chi_sq_alpha',chisq_alpha)

### Step 3 : Calculate Sample Statistic 
test_stat, p, dof, expected_value = stats.chi2_contingency(observed = observed_value, correction = False)
print("\nExpected values :\n", expected_value)
print("\n\nDegrees of freedom :", dof)
print("\nchisq :", test_stat)

### Step 4 : Critical value method
# Since chisq_alpha < chisq , the sample falls in rejection zone
# We donot reject Ho

### Step 5 : pvalue
print("\np-value :", p)
print('\nalpha :',alpha)
# Since pvalue < alpha , we reject Ho

### Step 6 : Communication of decision
# Hence, Observed frequency and Expected frequency does not have the same distribution.
```
